## Chi-square тест

Для выявления взаимосвязи между двумя номинативными переменными используется тест хи-квадрат (Chi-square Test). Напишем функцию, для расчета критерия хи-квадрат и p-value. 

На вход функция будет принимать список с номинативными переменными. На выходе получим датафрэйм с критерием хи-квадрат и p-value. Здесь нулевая гипотеза H0: переменные не взаимосвязаны. 

Если p_value < 0.05, отклоняем нулевую гипотезу - переменные не взаимосвязаны и принимаем альтернативную. Если p_value > 0.05, следовательно, принимаем нулевую гипотезу.

```python
lst_cat = ['Geography', 'Gender', 'Complain', 'Is_Active_Member', 'Has_CrCard', 'Num_Of_Products', 'Satisfaction_Score', 'Card_Type']
```

```python
def chi_square(lst):
    dictionary = {}
    dictionary['column'] = []
    dictionary['chi2'] = []
    dictionary['p_value'] = []
    for x in lst:
        contingency_table = pd.crosstab(df['Exited'], df[x], margins=True)
        chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)
        
        dictionary['column'].append(x)
        dictionary['chi2'].append(np.round(chi2, 2))
        dictionary['p_value'].append(np.round(p_value, 2))

    chi2_df = pd.DataFrame(dictionary)
    chi2_df['flag'] = chi2_df['p_value'].apply(lambda x: 1 if x < 0.05 else 0)
    return chi2_df
```

```python
chi_square(lst_cat)
```

Тест хи-квадрат (Chi-square Test) показал, что переменные Has_CrCard, Satisfaction_Score и Card_Type не связаны с переменной Exited.

## Тест Манна-Уитни

Для выявления взаимосвязи между номинативной и количественной переменной используется t-критерий Стьюдента. Он применяется при нормальном распределении случайной величины. Т.к. у нас не идеальное нормальное распределение и есть небольшие выбросы, то лучше использовать тест Манна-Уитни(Mann_Whitney Test), который не так чувствителен к выбросам как t-тест.

Напишем функцию, для расчета U-статистики и p-value. 

На вход функция будет принимать список с количественными переменными. На выходе получим датафрэйм со значением U-статистики и p-value. Здесь нулевая гипотеза H0: распределение случайной величины в обеих группах одинаково, т.е. переменные не взаимосвязаны. 

Если p_value < 0.05, отклоняем нулевую гипотезу и принимаем альтернативную. Если p_value > 0.05 принимаем нулевую гипотезу.

```python
lst_col = ['Credit_Score', 'Tenure', 'Age', 'Balance', 'Estimated_Salary', 'Point_Earned']
```

```python
def mann_whitney(lst):
    dictionary = {}
    dictionary['column'] = []
    dictionary['U-statistic'] = []
    dictionary['p_value'] = []
    for x in lst:
        statistic, p_value = stats.mannwhitneyu(df.query('Exited == 1')[x], df.query('Exited == 0')[x])
        
        dictionary['column'].append(x)
        dictionary['U-statistic'].append(np.round(statistic, 2))
        dictionary['p_value'].append(np.round(p_value, 2))

    mann_whitney_df = pd.DataFrame(dictionary)
    mann_whitney_df['flag'] = mann_whitney_df['p_value'].apply(lambda x: 1 if x < 0.05 else 0)
    return mann_whitney_df
```

```python
mann_whitney(lst_col)
```

Тест Манна-Уитни показал, что переменные Tenure, Estimated_Salary и Point_Earned не связаны с переменной Exited.

`Для анализа непрерывных переменных будем использовать функцию qcut. qcut - это функция, которая делит данные на заданное число интервалов, так чтобы в каждом интервале было примерно одинаковое количество наблюдений (если это возможно).`

## Balance

Помним, что клиентов с нулевым балансом на счету нужно выделить в отдельную группу

```python
# Функция на вход принимает столбец (df['Balance']) и n_col - число групп
def balance_cut(column, n_col):
    # Создаем отдельную серию для результатов
    result = pd.Series(index=column.index, dtype='object')
    # Для Balance = 0 будет отдельная группа 'Zero Balance'
    result[column == 0] = 'Zero Balance'

    # Для Balance > 0 используем qcut
    quantile_groups = pd.qcut(df[column > 0][column.name], q=n_col)
    result[column > 0] = quantile_groups

    
    return result
```

```python
df['Balance_interval'] = balance_cut(df['Balance'], n_col = 3)
```

```python
df.groupby('Balance_interval', observed=True)['Exited'].agg(['count', 'mean'])
```

```python
df.groupby('Balance_interval', observed=True)['Exited'].agg(['mean']).plot(kind = 'bar', rot=45, legend=False)
```

- Complain (есть ли у клиента жалоба). 99.5% клиентов, у которых была жалоба, ушли из банка. Т.е. если у клиента есть жалоба, то он скорее всего перестанет быть клиентом банка. При отсутствии жалоб вероятность ухода клиента очень низкая (0.05%). Компании следует обратить на это внимание.
- IsActiveMember (Активные клиенты). Статус активности клиента влияет на отток клиентов из банка. Неактивные клинты чаще покидают банк (26.8%) по сравнению с активными клиентами (14.3%).
- NumOfProducts (количество продуктов, купленных клиентом через банк). Влияет на отток клиентов из банка. Мы не можем сказать, что 3 и 4 купленных продукта сильно влияют на уход клиента, т.к. по ним очень мало данных (266 и 60 клиентов соответственно). Но можем сделать вывод, что доля ушедших клиентов с 1 продуктом (27.7%) выше, чем доля ушидших клиентов с двумя продуктами (7.6%).
- Age (возраст клиента). Возрастная группа клиента сильно влияет на показатель оттока. Из анализа видно, что с возрастом клиентов показатель оттока увеличивается. Это означает, что клиенты старшего возраста более склонны к уходу, в то время как молодые клиенты с большей вероятностью останутся в банке.
- Balance (Баланс на счетах клиента). Баланс на счетах клиента влияет на отток клиентов из банка. Так, клиенты с нулевым балансом реже покидают банк (13.8%), чем с высоким, средним и низким балансом.

```python
params = {'n_estimators': range(10, 101, 10),
          'max_depth': range(1, 22, 2),
          'min_samples_leaf': range(1, 8),
          'min_samples_split': range(2, 20, 2),
          'criterion': ['entropy', 'gini']}
```

```python
grid_search_cv_clf = RandomizedSearchCV(estimator=rf_clf,
                                        param_distributions=params,
                                        n_iter = 40,
                                        n_jobs = -1,
                                        cv=5, 
                                        verbose=2
                                        )
grid_search_cv_clf
```

```python
y_predict = best_clf.predict(X_test)
```

```python
y_predicted_proba = best_clf.predict_proba(X_test)[:, 1]
```

```python
def metrics(y_true, y_pred, y_pred_proba, model_name):
    lst = []
    lst.append(np.round(accuracy_score(y_true, y_pred), 4))
    lst.append(np.round(precision_score(y_true, y_pred), 4))
    lst.append(np.round(recall_score(y_true, y_pred), 4))
    lst.append(np.round(f1_score(y_true, y_pred), 4))
    lst.append(np.round(roc_auc_score(y_true, y_pred_proba), 4))
    t = pd.DataFrame({model_name:lst}, 
                     index = ['accuracy_score', 'precision_score', 'recall_score', 'f1_score', 'roc_auc_score']
                    )
    return t
```

```python
baseline_metrics = metrics(y_test, y_predict, y_predicted_proba, 'Baseline')
baseline_metrics
```

Метрика Recall составила 0.46. Т.к. мы хотим наиболее точно определять ушедших клиентов, то для нас это плохой результат. При этом получился довольно высокий Precision и неплохой для дисбаланса классов f1_score, равный 0.58.

```python
def conf_matrix_metrics(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(cm)
    disp.plot();

    specificity = cm[0,0]/(cm[0,0]+cm[0,1])
    print(f'specificity: {specificity:.2f}')
    print(f'fpr: {(1-specificity):.2f}')

    sensitivity = cm[1,1]/(cm[1,1]+cm[1,0])
    print(f'sensitivity: {sensitivity:.2f}')
    print(f'fnr: {(1-sensitivity):.2f}')
```

```python
conf_matrix_metrics(y_test, y_predict)
```

```python
fpr, tpr, thresholds = roc_curve(y_true=y_test, y_score=y_predicted_proba)
roc_auc= auc(fpr, tpr)
plt.plot(fpr, tpr, color='darkorange',
         label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--', label='random classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```

ROC-AUC составил 0.86. Это говорит о том, что модель достаточно хорошо разделяет классы. Но низкий Recall указывает на то, что порог 0.5 выбран неоптимально.

## Improved Model

Посмотрим, как можно улучшить модель, чтобы поднять Recall

### Делим на обучающую и тестовую выборку

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)
```

### Выделяем валидационную выборку

Валидационную выборку будем использовать для подбора порога решающего правила

```python
X_train_base, X_test_val, y_train_base, y_test_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42, stratify=y_train)
```

### Resampling

```python
df['Exited'].value_counts()
```

Наблюдается проблема дисбаланса классов, класс 0 (клиент остался) значительно преобладает над классом 1 (клиент ушел). В такой ситуации наша модель может просто "забить" на малочисленный класс, сосредоточившись на многочисленных классах.

Для балансировки классов воспользуемся методом SMOTE (Synthetic Minority Over-sampling Technique). SMOTE не просто дублирует минорные классы, как это делает простой oversampling, он создает новые, синтетические примеры, которые помогают модели лучше понять и обобщить характеристики минорных классов.

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy='auto', random_state=42)
# Применяем SMOTE только к обучающей выборке
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_base, y_train_base)
```

```python
y_train_resampled.value_counts()
```

### StandardScaler

StandardScaler приводит данные к стандартному виду, где среднее значение равно 0, а стандартное отклонение — 1 (Z-стандартизация: (x – mean) / std).

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Создание scaler
scaler = StandardScaler()
col_var=['Credit_Score', 'Age', 'Tenure', 'Balance', 'Estimated_Salary', 'Point_Earned']

for col in col_var:
    #обучение scaler только на train данных
    X_train_resampled[col] = scaler.fit_transform(X_train_resampled[[col]])

    # к test данным применяем transform
    X_test[col] = scaler.transform(X_test[[col]])
    X_test_val[col] = scaler.transform(X_test_val[[col]])
```

```python
X_train_resampled.head()
```

### Модель

Для борьбы с дисбалансом я уже сделал SMOTE, который создает новые синтетические образцы минорного класса (класса 1 в нашем случае). При создании объекта модели также укажу class_weight='balanced', что придает больший вес минорному классу.

```python
rf_clf = RandomForestClassifier(class_weight='balanced', random_state=42)
```

```python
params = {'n_estimators': range(10, 101, 10),
          'max_depth': range(1, 22, 2),
          'min_samples_leaf': range(1, 8),
          'min_samples_split': range(2, 20, 2),
          'criterion': ['entropy', 'gini']}
```

```python
grid_search_cv_clf = RandomizedSearchCV(estimator=rf_clf,
                                        param_distributions=params,
                                        n_iter = 40,
                                        n_jobs = -1,
                                        cv=5,
                                        verbose=2,
                                        scoring = 'recall') # оптимизируем метрику Recall
grid_search_cv_clf
```

```python
grid_search_cv_clf.fit(X_train_resampled, y_train_resampled)
```

```python
grid_search_cv_clf.best_score_
```

```python
grid_search_cv_clf.best_params_
```

```python
best_clf = grid_search_cv_clf.best_estimator_
```

### Подбор порога Recall

```python
# Получаем вероятности отнесения наблюдения к классу 1 для валидационной выборки
y_val_proba = best_clf.predict_proba(X_test_val)[:, 1]
y_val_proba
```

```python
# Пробуем разные пороги
thresholds = np.arange(0.1, 0.9, 0.05)
best_recall = 0
best_threshold = 0.5
best_precision_at_best_recall = 0

for thresh in thresholds:
    y_val_pred = (y_val_proba >= thresh).astype(int)
    recall_val = recall_score(y_true = y_test_val, y_pred = y_val_pred)
    precision_val = precision_score(y_true = y_test_val, y_pred = y_val_pred, zero_division=0)
    print(f'{thresh:.2f}  | {recall_val:.4f} | {precision_val:.4f}')
    
    if recall_val > best_recall:
        best_recall = recall_val
        best_threshold = thresh
        best_precision_at_best_recall = precision_val

print(f' Лучший порог: {best_threshold:.2f}')
print(f' Recall на валидации: {best_recall:.4f}')
print(f' Precision на валидации: {best_precision_at_best_recall:.4f}')
```

Допустим, меня устроит Recall равный 0.85. При таком Recall Precision будет равен 0.33.

```python
predict_val = (y_val_proba >=0.20).astype(int)
predict_val
```

```python
metrics(y_test_val, predict_val, y_val_proba, 'Validation')
```

```python
conf_matrix_metrics(y_test_val, predict_val)
```

```python
fpr, tpr, thresholds = roc_curve(y_true=y_test_val, y_score=y_val_proba)
roc_auc= auc(fpr, tpr)
plt.plot(fpr, tpr, color='darkorange',
         label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--', label='random classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```

ROC-AUC составил 0.80. Это хороший результат, который говорит о том, что модель достаточно хорошо разделяет классы.

Таким образом, на валидационных данных мы подобрали порог 0.20, который дает нам Recall = 0.85, f1_score = 0.48, ROC-AUC = 0.80. Остается протестировать порог на тестовых данных, которые модель еще не "видела".
